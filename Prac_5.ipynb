**Two Way F-test**

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# Load your dataset
df_anova2 = pd.read_csv("Sales_with_NaNs_v1.3.csv")
```

```python
# Two-way ANOVA: Sales_Before explained by Group and Customer_Segment
model = ols('Sales_Before ~ C(Group) * C(Customer_Segment)', data=df_anova2).fit()
print(f"Overall model F({model.df_model:.0f},{model.df_resid:.0f}) = {model.fvalue:.3f}, p = {model.f_pvalue:.4f}")
```

```python
# ANOVA table
res = sm.stats.anova_lm(model, typ=2)
res
```

Conclusion :

In two-way ANOVA, we reject the null hypothesis for a factor when the p-value is less than 0.05, meaning that factor (Group, Customer_Segment, or their interaction) has a significant effect on Sales_Before and the differences are unlikely due to chance. We fail to reject the null hypothesis when the p-value is greater than 0.05, meaning the data does not provide enough evidence that the factor affects sales. In simple terms, a clear and significant effect leads to rejecting H0, while no clear effect leads to accepting H0.

**Overall Conclusion :**

the one-way ANOVA only compares the mean Sales_Before across different Groups, so it checks just one factor. The two-way ANOVA, however, includes both Group and Customer_Segment plus their interaction, so it tests more effects at the same time. Because the two tests analyze different factors, their results will not be the same. One-way ANOVA gives a simple comparison, while two-way ANOVA gives a deeper explanation of how multiple factors together affect the sales.
