**t-test**

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
df = pd.read_csv("/content/Sales_with_NaNs_v1.3.csv")
```

```python
df.head()
```

```python
# Drop NaN values to avoid errors
data = df["Sales_Before"].dropna().values
print("Data:", data)
```

```python
# Define the null hypothesis
H0 = "The average Sales_Before is 32."
H1 = "The average Sales_Before is more than 32."
```

```python
# Calculate the test statistic
t_stat, p_value_two_tailed = stats.ttest_1samp(data, 32)
```

```python
# Convert two-tailed p-value to one-tailed (because your H1 is 'greater than')
p_value = p_value_two_tailed / 2
```

```python
# Print results
print("Test statistic:", t_stat)
print("One-tailed p-value:", p_value)
```

```python
# Conclusion
if t_stat > 0 and p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

Conclusion :

If the p-value is less than 0.05, we reject the null hypothesis as there is sufficient evidence that the data reflect a true difference and not just random due to chance. We fail to reject the null hypothesis if p-value > 0.05, which means that the data doesn’t provide enough evidence for any real difference. In brief, if the data is convincing then we say no to H0, and if not compelling enough we say yes.

**Two sampled T-test**

```python
from scipy.stats import ttest_ind
import pandas as pd
import numpy as np
```

```python
# Load dataset
df = pd.read_csv("Sales_with_NaNs_v1.3.csv")
```

```python
# Select groups (example: Control vs Treatment)
week1 = df[df["Group"] == "Control"]["Sales_Before"].dropna().values
week2 = df[df["Group"] == "Treatment"]["Sales_Before"].dropna().values

print("week1 data :-\n")
print(week1)

print("week2 data :-\n")
print(week2)
```

```python
# Means
week1_mean = np.mean(week1)
week2_mean = np.mean(week2)
print("week1 mean value:", week1_mean)
print("week2 mean value:", week2_mean)
```

```python
# Standard deviations
week1_std = np.std(week1)
week2_std = np.std(week2)
print("week1 std value:", week1_std)
print("week2 std value:", week2_std)
```

```python
# Independent t-test
ttest, pval = ttest_ind(week1, week2, equal_var=False)  # safer for unequal sample sizes
print("p-value:", pval)
```

```python
# Conclusion
if pval < 0.05:
    print("we reject null hypothesis")
else:
    print("we accept null hypothesis")
```

Conclusion :

Our null hypothesis is rejected when the p-value is less than 0.05, which means that the difference between two groups is statistically significant and there’s a low probability it was due to chance; i.e. these data support the TRUE treatment effect – not due to random variation in patients assigned to Treatment Group A or Group B. If the p-value is above 0.05, we will not reject the null hypothesis and indicate that becuase there isn't enough information in the data to prove a real difference. In brief: If the difference is big and clear, we reject H0; if it's small or unclear, we don't.

**Paired sampled t-test**

```python
import pandas as pd
from scipy import stats
```

```python
# Load your dataset
df = pd.read_csv("Sales_with_NaNs_v1.3.csv")
```

```python
# Describe Sales Before and After
print(df[['Sales_Before', 'Sales_After']].describe())
```

```python
# Paired t-test (Before vs After for same customers)
ttest, pval = stats.ttest_rel(df['Sales_Before'], df['Sales_After'], nan_policy='omit')

print("p-value:", pval)
```

```python
# Conclusion
if pval < 0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

Conclusion :

We reject the null hypothesis when the p-value is less than 0.05. This means that the difference between Sales_Before and Sales_After is statistically significant and unlikely to have happened by chance. On the other hand, if the p-value is greater than 0.05, we don’t have enough evidence to reject the null hypothesis, suggesting that any change in sales isn’t clearly significant. Simply put, a noticeable and meaningful change in sales leads us to reject H0, while a small or unclear change means we keep H0.



**Z test**

You would use a Z test if:

Your sample size is greater than 30. Otherwise, use a t test.

Data points should be independent from each other. In other words, one data point isn’t related or doesn’t affect another data point.

Your data should be normally distributed. However, for large sample sizes (over 30) this doesn’t always matter.

Your data should be randomly selected from a population, where each item has an equal chance of being selected.

Sample sizes should be equal if at all possible.

```python
import pandas as pd
from scipy import stats
from statsmodels.stats import weightstats as stests
```

```python
# Load dataset (only if not already loaded)
df = pd.read_csv("Sales_with_NaNs_v1.3.csv")
```

```python
# Run one-sample Z-test on Sales_Before
ztest, pval = stests.ztest(df['Sales_Before'].dropna(),
                           x2=None,
                           value=156)  # hypothesized mean

print(float(pval))
```

```python
# Conclusion
if pval < 0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

Conclusion :

We reject the null hypothesis when the p-value is less than 0.05. This means that the average Sales_Before is significantly different from the hypothesized value and that this difference is unlikely to be due to chance. If the p-value is greater than 0.05, we fail to reject the null hypothesis, indicating that the data doesn’t provide enough evidence to confirm a real difference. In simple terms, a clear and noticeable difference from the hypothesized mean leads us to reject H0, while a small or uncertain difference means we keep H0.

**Two-sample Z test**

```python
import pandas as pd
from statsmodels.stats import weightstats as stests
```

```python
# Load dataset
df = pd.read_csv("Sales_with_NaNs_v1.3.csv")
```

```python
# Paired Z-test: Sales_Before vs Sales_After
ztest, pval1 = stests.ztest(df['Sales_Before'].dropna(),
                            x2=df['Sales_After'].dropna(),
                            value=0,
                            alternative='two-sided')

print(float(pval1))
```

```python
# Conclusion
if pval1 < 0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

Conclusion :

We reject the null hypothesis when the p-value is less than 0.05, which means the difference between Sales_Before and Sales_After is statistically significant and unlikely to be due to chance. When the p-value is greater than 0.05, we fail to reject the null hypothesis, indicating that the data doesn’t provide enough evidence of a real change. In simple terms, a clear and meaningful change in sales leads us to reject H0, while a small or uncertain change means we retain H0.

**Chi-Square Test**

```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
# Load your dataset
df = pd.read_csv('Sales_with_NaNs_v1.3.csv')
```

```python
# Create a contingency table
contingency_table = pd.crosstab(df['Group'], df['Purchase_Made'])
```

```python
# Perform the chi-square test
chi2_statistic, p_value, dof, expected_frequencies = chi2_contingency(contingency_table)
```

```python
# Print the results
print('Chi-square statistic:', chi2_statistic)
print('P-value:', p_value)
print('Degrees of freedom:', dof)
print('Expected frequencies:')
print(expected_frequencies)
```

Conclusion :

We reject the null hypothesis when the p-value is less than 0.05, which means there is a significant association between Group and Purchase_Made, and the observed differences are unlikely to be due to chance. When the p-value is greater than 0.05, we fail to reject the null hypothesis, indicating that the data doesn’t provide enough evidence of a real relationship. In simple terms, a clear association leads us to reject H0, while no clear association means we retain H0.

**Overall Conclusion :**

The choice of test depends on the type of data and comparison: use one-sample t-test or Z-test to compare a single group’s mean to a known value, two-sample t-test or Z-test to compare two independent groups, and paired t-test or paired Z-test to compare before-and-after values for the same customers. Use Chi-Square test to check associations between categorical variables like Group and Purchase_Made. For analyzing multiple factors together, ANOVA (especially two-way ANOVA) is best, as it shows the effects of Group, Customer_Segment, and their interaction on sales.
